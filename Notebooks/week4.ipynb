```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/pre_proccesed_data.csv")
df
```

```python
X = df.drop(columns=['cardio'])  
y = df['cardio']
```

```python
X = X.values
y = y.values.reshape(-1, 1)
```

```python
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)

X = (X - X_mean) / X_std
```

```python
m = X.shape[0]
X = np.c_[np.ones((m, 1)), X]
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def compute_cost(X, y, weights):
    m = len(y)
    h = sigmoid(X @ weights)
    cost = -(1/m) * np.sum(y*np.log(h + 1e-9) + (1-y)*np.log(1-h + 1e-9))
    return cost
```

```python
def gradient_descent(X, y, weights, lr, iterations):
    m = len(y)
    cost_history = []

    for _ in range(iterations):
        h = sigmoid(X @ weights)
        gradient = (1/m) * (X.T @ (h - y))
        weights -= lr * gradient
        cost_history.append(compute_cost(X, y, weights))

    return weights, cost_history
```

```python
weights = np.zeros((X.shape[1], 1))
learning_rate = 0.01
iterations = 1000
```

```python
weights, cost_history = gradient_descent(
    X, y, weights, learning_rate, iterations
)
```

```python
def predict(X, weights):
    probs = sigmoid(X @ weights)
    return (probs >= 0.5).astype(int)
```

```python
y_pred = predict(X, weights)
```

```python
accuracy = np.mean(y_pred == y) * 100
print("Model Accuracy:", accuracy)
```
